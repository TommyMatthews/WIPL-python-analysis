```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Load the CSV file
coord_file = "three_bernard_coords_10_20_100_30.csv"
drive_path_to_coords = BASE_DRIVE_PATH = '/Users/sstk4353/Library/CloudStorage/GoogleDrive-thomas.matthews@ncas.ac.uk/My Drive/WIPL/Bernard_ellipsoid_comparison/multi_moth_outputs/coordinates/'
df_coords = pd.read_csv(drive_path_to_coords + coord_file)

# Extract coordinates for Body 2 and Body 3
x2, y2, z2 = df_coords['x2'], df_coords['y2'], df_coords['z2']
x3, y3, z3 = df_coords['x3'], df_coords['y3'], df_coords['z3']

a, b = df_coords.shape

print(a, b)
```

```python
# Create 3D scatter plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot Body 1 (origin)
ax.scatter(0, 0, 0, color='black', s=80, label='Body 1 (Origin)', marker='o')

# Plot Body 2 and Body 3
ax.scatter(x2, y2, z2, color='blue', label='Body 2', alpha=0.7)
ax.scatter(x3, y3, z3, color='red', label='Body 3', alpha=0.7)

# Labeling
ax.set_xlabel('X (mm)')
ax.set_ylabel('Y (mm)')
ax.set_zlabel('Z (mm)')
ax.set_title('3D Locations of Three Bodies (mm)')
ax.legend()
ax.grid(True)

plt.tight_layout()
plt.show()
```
